```python
import matplotlib
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
time = [20070101, 20070401, 20070701, 20071001, 20080101, 20080401, 20080701, 20081001, 20090101]
time
```

```python
plt.plot(time)
plt.xlabel('Measurement #')
plt.ylabel('YearMonthDay')
```

```python
gdp = [14233.2, 14422.3, 14569.7, 14685.3, 14668.4, 14813.0, 14843.0, 14549.9, 14383.9]
gdp
```

```python
plt.plot(time,gdp,'o-')
plt.xlabel('Time in formation YearMonthDay')
plt.ylabel('GDP in Billions')
```

```python
from datetime import datetime
```

```python
date = []
for t in time: #for each element time
    #print(t)
    date.append(datetime.strptime(str(t),'%Y%m%d'))
    # %Y means 4 character year
    # %m means two character month
    # %d means two chracter date
```

```python
print(date)
```

```python
time
```

```python
plt.plot(date,gdp,'o-')
plt.xlabel('Time in formation YearMonthDay')
plt.ylabel('GDP in Billions')
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(date,gdp,'o-')
plt.show() # this ignores the memory address created in the previous step while displaying the graph
```

```python
style = 'fivethirtyeight'  #make plot in fivethirtyeight style
with plt.style.context(style):
    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(date,gdp,'o-')
    plt.show() 
```

```python
# into a function
def make_gdp_plot(style):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(8,5))
        ax.plot(date,gdp,'o-')
        plt.show() 
```

```python
make_gdp_plot('fivethirtyeight')
```

```python
make_gdp_plot('ggplot')
```
